```
import pandas as pd
dataset = pd.read_csv("forestfires.csv")
```

```
area_dataset = dataset[dataset['area'] > 0]
print(area_dataset['area'].mean())
print(area_dataset['area'].min())
print(area_dataset['area'].max())
print(area_dataset['area'].std())
print(area_dataset.sort_values(by=['area']).tail(20))
print(area_dataset['area'].median())
print(area_dataset['month'].unique())
print(dataset['month'].unique())
# print(dataset[dataset['month'] == 'mar'])
print(dataset[dataset['month'] == 'mar'].shape)
print(dataset[dataset['month'] == 'mar'].shape[0])
```

```
months = dataset["month"].unique()
for month in months:
  month_dataset = dataset[dataset['month'] == month]
  fires_in_month = month_dataset.shape[0]
  avg_tmp_in_month = int(month_dataset['temp'].mean())

  print(str(fires_in_month) + " fires in " + month + " with a mean temperature of " + str(avg_tmp_in_month) + "Â°C")
```
