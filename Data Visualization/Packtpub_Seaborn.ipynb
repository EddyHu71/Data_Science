```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```
sns.set(style="ticks")
data = pd.read_csv("salary.csv")[:500]
sns.relplot(x="Salary", y="Age", hue="Education", style="Education", col="Gender", data=data)
```

```
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label="Group A")
plt.plot(x2, label="Group B")
plt.legend()
plt.show()
```
